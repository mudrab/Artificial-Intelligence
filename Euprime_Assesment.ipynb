```python
import pandas as pd

class LeadGenAgent:
    def __init__(self):
        self.weights = {
            "scientific_intent": 40,
            "role_fit": 30,
            "funding_intent": 20,
            "location_hub": 10
        }

    def calculate_score(self, lead):
        score = 0
        if lead['has_recent_pub']: score += self.weights['scientific_intent']
        if "Toxicology" in lead['title'] or "Safety" in lead['title']: score += self.weights['role_fit']
        if lead['recent_funding']: score += self.weights['funding_intent']
        if lead['is_in_hub']: score += self.weights['location_hub']
        return score

    def process_leads(self, raw_data):
        df = pd.DataFrame(raw_data)
        df['probability_score'] = df.apply(self.calculate_score, axis=1)
        return df.sort_values(by='probability_score', ascending=False)

# Mock Data representing Stage 1 & 2 (Identification & Enrichment)
data = [
    {
        "name": "Dr. Jane Doe",
        "title": "Director of Toxicology",
        "company": "BioPharma Corp",
        "has_recent_pub": True,
        "recent_funding": True,
        "is_in_hub": True,
        "location": "Boston, MA",
        "hq": "Boston, MA"
    },
    {
        "name": "John Smith",
        "title": "Junior Researcher",
        "company": "Startup X",
        "has_recent_pub": False,
        "recent_funding": False,
        "is_in_hub": False,
        "location": "Remote",
        "hq": "Austin, TX"
    }
]

agent = LeadGenAgent()
qualified_leads = agent.process_leads(data)
print(qualified_leads)
# qualified_leads.to_csv("output_for_google_sheets.csv")
```
